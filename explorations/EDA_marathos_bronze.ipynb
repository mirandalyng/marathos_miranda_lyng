## EDA - Marathos (Ultra Marathon Dataset)

### Basic EDA before ingesting into bronze layer


```python
VOLUME_PATH = "/Volumes/marathos/default/raw"

spark.sql(f"LIST '{VOLUME_PATH}'").display()

spark.sql(f"LIST '{VOLUME_PATH}/data'").display()
```

```python
%sql 
SHOW SCHEMAS IN marathos; 
```

```python
df = (
    spark.read.format("csv")
    .options(header=True, inferSchema=True)
    .load(f"{VOLUME_PATH}/data/TWO_CENTURIES_OF_UM_RACES.csv")
    #used LMM to remove .schema here to be able get the dataframe
) 

df
```

```python
df.printSchema()
```

```python
df.columns
```

```python
len(df.columns)
```

```python
df.count()
```

```python
df.describe().display()
```

```python
for col in df.columns:
    print(col, df.filter(df[col].isNull()).count())
```

```python
df.select("Athlete club").display()


null_count = df.filter(df["Athlete club"].isNull()).count()

print(null_count)
    
```

```python
df.select("Event name").distinct().count()
```

```python
df.select("Event dates").distinct().count()
```

```python
df.select("Event name", "Event dates").distinct().display()
```

```python
df.select("Event dates").display()
```

```python
df.select("Athlete gender").display()

#count M & F athletes 

Male = df.filter(df["Athlete gender"] == "M").count()
Female = df.filter(df["Athlete gender"] == "F").count()

print(f"Male: {Male}, Female: {Female}")


```

```python
df.select("Athlete age category").distinct().display()
```

```python
df.groupBy("Athlete gender", "Athlete age category").count().display()
```

```python

df.filter(df["Athlete country"] == "USA").count()
```

```python
df.groupBy("Athlete country").count().orderBy("count", ascending=False).limit(10).display()
```

```python
df.select("Year of event").distinct().orderBy("Year of event",ascending=True).display()
```
