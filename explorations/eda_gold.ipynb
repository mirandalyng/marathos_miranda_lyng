```python
%sql
SELECT 
  fr.result_id,
  fr.athlete_performance_time_h,
  e.event_unit,
  e.event_id
FROM marathos.gold.fct_results fr
LEFT JOIN marathos.gold.dim_event e ON fr.event_id = e.event_id
LIMIT 20;
```

```python
%sql
SELECT event_unit, COUNT(*) 
FROM marathos.gold.dim_event 
GROUP BY event_unit;
```

```python
%sql
SELECT 
  fr.event_id,
  e.event_id,
  e.event_unit,
  fr.athlete_performance_time_h
FROM marathos.gold.fct_results fr
LEFT JOIN marathos.gold.dim_event e ON fr.event_id = e.event_id
WHERE e.event_unit = 'h'
LIMIT 20;
```

```python
%sql
SELECT 
  event_id,
  athlete_performance_time_h
FROM marathos.gold.fct_results
WHERE athlete_performance_time_h IS NOT NULL
LIMIT 10;
```

```python
%sql
SELECT 
  fr.event_id        AS fct_event_id,
  e.event_id         AS dim_event_id,
  fr.athlete_performance_time_h
FROM marathos.gold.fct_results fr
LEFT JOIN marathos.gold.dim_event e ON fr.event_id = e.event_id
LIMIT 10;
```
